```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

# Open the file:
raster_B02 = gdal.Open(r"D:\01_Documents\01_University\GEOL_495\04_Scripts\Data\Band_02.tif")
raster_B03 = gdal.Open(r"D:\01_Documents\01_University\GEOL_495\04_Scripts\Data\Band_03.tif")
raster_B04 = gdal.Open(r"D:\01_Documents\01_University\GEOL_495\04_Scripts\Data\Band_04.tif")

rasterArray_B02 = raster_B02.ReadAsArray()
rasterArray_B03 = raster_B03.ReadAsArray()
rasterArray_B04 = raster_B04.ReadAsArray()

myx = np.linspace(0, raster_B02.RasterXSize, raster_B02.RasterXSize)
plt.rcParams['figure.figsize'] = (10,10)
f, axarr = plt.subplots(3, sharex=True)
axarr[0].set_title('Sharing X axis')

axarr[0].plot(myx, rasterArray_B02[0])
axarr[1].plot(myx, rasterArray_B03[0])
axarr[2].plot(myx, rasterArray_B04[0])

for x in range(0, 3):
    axarr[x].grid()
plt.show()
```
